```python
import pandas as pd
df = pd.read_csv("COVID-19 India Statewise Vaccine Data.csv")
df.head()
```

```python
df.info()
df.columns
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Top 10 states by total vaccination
top_states = df.sort_values("Total Vaccination Doses", ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(data=top_states, x="Total Vaccination Doses", y="State/UTs", palette="viridis")
plt.title("Top 10 States by COVID-19 Vaccinations")
plt.xlabel("Total Vaccination Doses")
plt.ylabel("State/UTs")
plt.show()
```

```python
# Compare Dose1 and Dose2 for each state
plt.figure(figsize=(14,7))
df_sorted = df.sort_values("Dose1", ascending=False)

plt.bar(df_sorted["State/UTs"], df_sorted["Dose1"], label="Dose1", color="red")
plt.bar(df_sorted["State/UTs"], df_sorted["Dose 2"], label="Dose 2", color="black", alpha=0.7)

plt.xticks(rotation=90)
plt.xlabel("States/UTs")
plt.ylabel("Number of Doses")
plt.title("Dose1 vs Dose2 Vaccinations by State")
plt.legend()
plt.show()
```

```python
# Calculate vaccination % of population
df["Vaccination %"] = (df["Total Vaccination Doses"] / df["Population"]) * 100

# Plot Top 10 states by % vaccination
top_vaccination = df.sort_values("Vaccination %", ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(data=top_vaccination, x="Vaccination %", y="State/UTs", palette="coolwarm")
plt.title("Top 10 States by Vaccination % of Population")
plt.xlabel("Vaccination %")
plt.ylabel("State/UTs")
plt.show()
```

```python
# Booster doses by state
plt.figure(figsize=(12,6))
df_booster = df.sort_values("Precaution 18-59", ascending=False).head(10)

sns.barplot(data=df_booster, x="Precaution 18-59", y="State/UTs", palette="mako")
plt.title("Top 10 States by Precaution (Booster) Doses")
plt.xlabel("Precaution Doses (18-59)")
plt.ylabel("State/UTs")
plt.show()
```

# COVID-19 Vaccination Analysis in India  

This project analyzes the state-wise COVID-19 vaccination data in India.  
We use Python with Pandas and Matplotlib to clean, explore, and visualize the dataset.  

The goal is to understand vaccination distribution across states, doses given, and age group trends.

## Step 1: Import Libraries  

We start by importing the necessary Python libraries.  

- **Pandas**  for handling and analyzing the dataset.  
- **Matplotlib**  for creating charts and visualizations.  

## Step 2: Load the Dataset  

We load the CSV file containing the vaccination data.  
This dataset has information such as:  
- State/UT names  
- Total doses given  
- Dose 1 and Dose 2  
- Precaution doses  
- Vaccination by age groups (1518, 1845, 4560, 60+).  

## Step 3: Explore the Data  

Before analysis, we quickly check the first few rows and dataset structure.  
This helps us understand the available columns and verify that the data loaded correctly.

## Step 4: Data Cleaning  

We clean the dataset by:  
- Renaming columns for clarity  
- Handling missing or inconsistent values (if any)  
- Making sure numbers are in the correct format  

Clean data ensures that our visualizations and insights are accurate.

## Step 5: Total Vaccination by State  

Here, we visualize the total number of vaccination doses administered in each state.  
This helps us identify which states had the highest and lowest vaccination coverage.  

## Step 6: Dose 1 vs Dose 2 Comparison  

This chart compares the number of people who took Dose 1 and Dose 2 in each state.  
It highlights the vaccination completion rate across states.  

## Step 7: Precaution Dose Analysis  

We visualize the precaution doses given to different age groups.  
This is important because precaution doses were targeted at vulnerable populations like the elderly (60+).  

## Step 8: Vaccination by Age Group  

This chart shows how vaccinations were distributed among different age groups (1518, 1845, 4560, 60+).  
It gives insights into which age groups were prioritized and how coverage varied.  

## Step 9: Insights and Conclusion  

From the analysis, we observe that:  
- States with higher populations had larger vaccination numbers.  
- There is a noticeable difference between Dose 1 and Dose 2 coverage.  
- Precaution doses were mainly focused on older age groups.  
- The 1845 age group received the highest number of doses, showing mass vaccination efforts for the working-age population.
